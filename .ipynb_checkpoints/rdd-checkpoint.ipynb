```python
inputRDD = sc.textFile("example_log.log")
errorRDD = inputRDD.filter(lambda x: "ERROR"in x.split())
warningsRDD = inputRDD.filter(lambda x: "INFO"in x.split())
combinedRDD = errorRDD.union(warningsRDD)
```

```python
warningsRDD.count()
warningsRDD.first()
```

```python
stock = inputRDD.filter(lambda x: "INFO" in x.split())
stock.first()
```

```python

shares = warningsRDD.filter((lambda x: "{logging_mixin.py:109}" in x))
shares.collect()
```

```python
historical_shares = []
for info in shares.collect():
    share = info.split("INFO - ")
    if not "TaskInstance" in share[1]:
        historical_shares.append(share[1])
```

```python
historical_shares[1] == historical_shares[0]
```
